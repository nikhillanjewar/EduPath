```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
import json
# Load the sample data
csv_filename = "New_data.csv"
data = pd.read_csv(csv_filename)
```

```python
data.info()
```

```python
data.head()
```

```python
# Split the data into features (X) and labels (y)
X = data.iloc[:, :-1].values
# X
```

```python
y = data.iloc[:, -1].values
# y
```

```python
# Use LabelEncoder to encode categorical labels to numerical values
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
y_encoded
```

```python
label_encoder.inverse_transform(y_encoded)
```

**SAVE OUTPUT LABEL ENCODING**

```python
max(y_encoded)
```

```python
label_encoding_dict = {}
for i in range(max(y_encoded)+1):
    Output_lables=label_encoder.inverse_transform([i])
    decoded_label=Output_lables[0]
    # print(decoded_label)

       # Add the mapping to the dictionary
    label_encoding_dict[str(i)] = decoded_label
```

```python
label_encoding_dict
```

```python
# Save label encodings to a JSON file
with open('Output_label_encodings.json', 'w') as json_file:
    json.dump(label_encoding_dict, json_file)
```

---------------------------------------------

**ENCODE INPUT**

```python
# Use OneHotEncoder to one-hot encode categorical features
onehot_encoder = OneHotEncoder(sparse=False)
X_encoded = onehot_encoder.fit_transform(X)
X_encoded
```

```python
X_encoded.shape
```

```python
len(X_encoded[0])
```

**SAVE ONEHOT ENCODER**

```python
# Store the feature names and indices mapping
feature_indices_mapping = {str(i): feature_name for i, feature_name in enumerate(data.columns[:-1])}
```

```python
# Save the categories and feature indices mapping to a JSON file
encoding_info = {
    'categories': {str(i): cats.tolist() for i, cats in enumerate(onehot_encoder.categories_)},
    'feature_indices_mapping': feature_indices_mapping,
}

with open('onehot_encoder_encoding.json', 'w') as json_file:
    json.dump(encoding_info, json_file)
```

--------------------------------------------------------------------

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, to_categorical(y_encoded), test_size=0.2, random_state=42)
```

```python
# Build a simple neural network model
model = Sequential()
```

```python
# Input layer
model.add(Dense(256, input_dim=X_train.shape[1], activation='relu'))
# Additional hidden layers with dropout for regularization
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))  # Adding dropout for regularization
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.1))  # Adding dropout for regularization
```

```python



# model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
# model.add(Dense(32, activation='relu'))  # Additional hidden layer
# model.add(Dense(16, activation='relu'))  # Additional hidden layer
```

```python
model.add(Dense(to_categorical(y_encoded).shape[1], activation='softmax'))  # Output layer with dynamic number of units
# model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
# Convert the model to TensorFlow Lite format
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Save the TensorFlow Lite model to a file
tflite_filename = "model_final.tflite"
with open(tflite_filename, 'wb') as f:
    f.write(tflite_model)
```
